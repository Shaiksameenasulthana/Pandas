```python
# 21. Convert Shipping Cost, Profit, and Sales into Rupees, Yens, Pounds
conversion_rates = {
    'Rupees': 74.85,  # Example conversion rate
    'Yens': 110.57,   # Example conversion rate
    'Pounds': 0.73    # Example conversion rate
}

# Add converted columns for Shipping Cost
df['Shipping Cost (Rupees)'] = df['Shipping Cost'] * conversion_rates['Rupees']
df['Shipping Cost (Yens)'] = df['Shipping Cost'] * conversion_rates['Yens']
df['Shipping Cost (Pounds)'] = df['Shipping Cost'] * conversion_rates['Pounds']

# Add converted columns for Profit
df['Profit (Rupees)'] = df['Profit'] * conversion_rates['Rupees']
df['Profit (Yens)'] = df['Profit'] * conversion_rates['Yens']
df['Profit (Pounds)'] = df['Profit'] * conversion_rates['Pounds']

# Add converted columns for Sales
df['Sales (Rupees)'] = df['Sales'] * conversion_rates['Rupees']
df['Sales (Yens)'] = df['Sales'] * conversion_rates['Yens']
df['Sales (Pounds)'] = df['Sales'] * conversion_rates['Pounds']

```
